```python
import numpy as np
import matplotlib.pyplot as plt
```

$$M = \sigma\sqrt{2\pi}exp(\frac{\pi^2}{2\sigma^2} + 1)$$

```python
sigma2 = 1
M = np.sqrt(2*np.pi*sigma2)*np.exp(np.pi**2/(2*sigma2)+1)
# M = 360
print(M)
def pbar(x):
  return np.exp(np.sin(x))
x = []
i = 0
j = 1
while(i<10000):
  z = np.random.normal(0, 1**2)
  u = np.random.uniform(0,M*np.exp(-(z**2/2))/(np.sqrt(2 * np.pi)))
  if u< pbar(z) and z<=np.pi and z>=-np.pi: 
    x.append(z)
    i+=1
  j+=1
print('number of samples rejected = ',j-i)
plt.hist(x, 100, density = False) 
plt.show()
```
